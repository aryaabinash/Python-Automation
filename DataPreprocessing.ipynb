```python
import os
import numpy as np
import pandas as pd
from IPython.display import Image
os.getcwd()
```

```python
students = [('Vikas', 25, 'India'),
            ('Vikas',35,'Syden'),
            ('Anand',16,'Singapore'),
            ('Steve',30,'London'),
            ('Swati',20,'Delhi'),
            ('Anand',16,'Singapore'),
             ('Vikas',35,'Syden')]

df = pd.DataFrame(students, columns = ['Name','Age','City'])
df
```

```python
df.duplicated().sum()
```

```python
# Find duplicate rows based on all columns
```

```python
duplicateRowDF = df[df.duplicated(keep='first')]
print("Duplicate Rows except first occurance on all columns are:")
duplicateRowDF
```

```python
duplicateRowDF = df[df.duplicated(keep='last')]
print("Duplicate Rows except first occurance on all columns are:")
duplicateRowDF
```

```python
duplicateRowDF = df[df.duplicated()]
duplicateRowDF
```

```python
df[df.duplicated(['Name'],keep='last')]
```

```python
titanic = pd.read_csv('titanic_dataset.csv')
titanic.duplicated().sum()
```

```python
titanic.shape
```

**How to drop duplicate data**

```python
titanic.drop_duplicates(keep='first',inplace=True)
```

```python
titanic.shape
```

**Missing data treatment**

```python
titanic.isnull().sum()
```

```python
titanic.isnull().sum().sum()
```

```python
titanic.isnull().sum()/len(titanic)*100
```

Cabin has 77 percent missing value - 77.1% missing data found


```python
titanic = titanic.drop(['Cabin'],axis=1)
# axis = 1 means column wise drop
# axis = 0 means row wise drop
# When there is object no need of writing inplace=True
```
